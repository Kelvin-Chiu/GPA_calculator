```python
import numpy as np
import pandas as pd

class gpa_calculator:

    def __init__(self):
        # super.__init__()
        self.dic = None
        self.total_credits = None
        self.avg_credit_grades = None
        self.converted_table = None
        self.origin_table = None
        self.grade_points = None
        self.grade_intervals = None
        self.method = None
        self.table = None

    def set_table(self, filename, sheet_name):
        self.table = pd.read_excel(filename, sheet_name=sheet_name, header=0)
        return print('set_table done')

    def set_method(self, method):
        self.method = method
        return print('set_method done')

    def to_float(self, credit):
        if isinstance(credit, (int, float)):
            pass
        return float(credit)

    def cat2num(self, grade):
        if isinstance(grade, (int, float)):
            pass
        else:
            self.dic = {'优':95, '良':85, '中':75, '合格':80}
            return self.dic.get(grade)
        return int(grade)

    def get_cleaned_col(self, col, object):
        if object == 'grades':
            return col.apply(self.cat2num)
        elif object == 'credits':
            return col.apply(self.to_float)
        return col

    def get_measure(self):
        if self.method == '标准4.0':
            self.grade_intervals = [90, 80, 70, 60, 0]
            self.grade_points = [4.0, 3.0, 2.0, 1.0, 0.0]
        elif self.method == '改进4.0(1)':
            self.grade_intervals = [85, 70, 60, 0]
            self.grade_points = [4.0, 3.0, 2.0, 0.0]
        elif self.method == '改进4.0(2)':
            self.grade_intervals = [85, 75, 60, 0]
            self.grade_points = [4.0, 3.0, 2.0, 0.0]
        elif self.method == '北大4.0':
            self.grade_intervals = [90, 85, 82, 78, 75, 72, 68, 64, 60, 0]
            self.grade_points = [4.0, 3.7, 3.3, 3.0, 2.7, 2.3, 2.0, 1.5, 1.0, 0.0]
        elif self.method == '加拿大4.3':
            self.grade_intervals = [90, 85, 80, 75, 70, 65, 60, 0]
            self.grade_points = [4.3, 4.0, 3.7, 3.3, 3.0, 2.7, 2.3, 0.0]
        elif self.method == '中科大4.3':
            self.grade_intervals = [95, 90, 85, 82, 78, 75, 72, 68, 65, 64, 61, 60, 0]
            self.grade_points = [4.3, 4.0, 3.7, 3.3, 3.0, 2.7, 2.3, 2.0, 1.7, 1.5, 1.3, 1.0, 0.0]
        elif self.method == '上海交大4.3':
            self.grade_intervals = [95, 90, 85, 80, 75, 70, 67, 65, 62, 60, 0]
            self.grade_points = [4.3, 4.0, 3.7, 3.3, 3.0, 2.7, 2.3, 2.0, 1.7, 1.0, 0.0]
        elif self.method == '东大4.0':
            self.grade_intervals = [86, 83, 80, 75, 70, 65, 60, 0]
            self.grade_points = [4.0, 3.7, 3.3, 3.0, 2.7, 2.3, 2.0, 0.0]
        return print(f'get_measure done : {self.method}')

    def grades_convert(self, grade):
        for i in range(len(self.grade_intervals)):
            if grade >= self.grade_intervals[i]:
                return self.grade_points[i]

    def get_calculate_result(self):
        self.origin_table = self.table.iloc[:]
        self.converted_table = self.table.iloc[:]
        self.converted_table['整理后成绩'] = self.converted_table['最终'].apply(self.grades_convert)
        self.converted_table['转换后绩点'] = self.converted_table.apply(lambda row: row['学分'] * row['整理后成绩'], axis=1)
        print('转换完成，开始计算')
        self.total_credits = self.converted_table['学分'].sum()
        self.avg_credit_grades = self.converted_table['转换后绩点'].sum()/self.total_credits
        return print('calculate done')

    def run(self):
        self.table['最终'] = self.get_cleaned_col(self.table['最终'], object='grades')
        self.table['学分'] = self.get_cleaned_col(self.table['学分'], object='credits')
        self.get_measure()
        self.get_calculate_result()
        print(f'avg_credit_grades : {self.avg_credit_grades}')
```

```python
# 可选算法有：标准4.0、改进4.0(1)、改进4.0(2)、北大4.0、加拿大4.3、中科大4.3、上海交大4.3、东大4.0

my_cal = gpa_calculator()
my_cal.set_table('./test.xlsx', 0) # test.xlsx是测试用的成绩表，0是sheet名，可自行修改
my_cal.set_method('东大4.0') # 算法可自行修改
my_cal.run()
```
